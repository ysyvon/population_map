```python
import os

# Get the current directory (where the notebook is located)
current_dir = os.getcwd()

# Construct the correct path to the shapefile in the 'data' folder (relative path)
shapefile_path = os.path.join(current_dir, "..", "data", "ne_110m_populated_places_simple.shp")

# Print the constructed path for debugging
print(f"Shapefile path: {shapefile_path}")

# Load the shapefile
import geopandas as gpd
gdf = gpd.read_file(shapefile_path)

# Check if the data loaded correctly
print(gdf.head())
```

```python
# Check the CRS
print(gdf.crs)
```

```python
# Inspect the Columns
print(gdf.columns)
```

```python
# Summarise the Data
print(gdf.info())
```

```python
# Check the first few rows
print(gdf.head())
```

```python
# Summarise Numeric Columns
print(gdf.describe())
```

```python
# List Unique Values
print(gdf['name'].unique())
```

```python
# Visualise all points
gdf.plot(figsize=(10, 10), color='blue', markersize=10)
```

```python
# Colour by population
gdf.plot(column='pop_max', cmap='viridis', legend=True, figsize=(12, 8))
```

```python
# Save as GeoJSON in the 'outputs' folder
# Ensure the outputs directory exists
output_dir = "outputs"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Save as GeoJSON in the 'outputs' folder
gdf.set_geometry("geometry", inplace=True)
gdf.to_file(os.path.join(output_dir, "populated_places_points.geojson"), driver="GeoJSON")
```

```python
import folium

# Create a base map
m = folium.Map(location=[20, 0], zoom_start=2)

# Add markers for each city
for _, row in gdf.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        popup=f"{row['pop_max']} ({row['name']})"
    ).add_to(m)

# Save and display the map
m.save("populated_places_map.html")
```
